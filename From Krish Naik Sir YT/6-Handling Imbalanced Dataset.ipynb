## Handling Imbalanced Dataset

- pip install imbalanced-learn

```python
import pandas as pd
df = pd.read_csv('Datasets/creditcard.csv')
df.head()
```

```python
df.shape
```

```python
df['Class'].value_counts()
```

```python
## Independent and Dependent Features
X = df.drop("Class",axis=1)
y = df.Class 
```

### When Dealing with Imbalanced Dataset go with using *Cross Validation* like *KFold* and *Hyperparameter Tuning*

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import KFold
import numpy as np 
from sklearn.model_selection import GridSearchCV
```

```python
logistic_regression = LogisticRegression()
grid = {
    'C': 10.0**np.arange(-2,3),
    'penalty': ['l1','l2'],     
}

cv = KFold(n_splits=5,random_state=None,shuffle=False)

```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
clf = GridSearchCV(logistic_regression,grid,cv=cv,n_jobs=-1,scoring='f1_macro')
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
## trying RandomForest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

It is performing better than LogisticRegression

```python
class_weight = dict({
    0:1,
    1:100
})
```

```python
## trying RandomForest with Class Weight
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(class_weight=class_weight)
rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

- When we have high imbalanced dataset then we can use the concept of class weight.
- Try to use this way at last.

#### Under Sampling
- Undersampling -> reduce the points of maximum labels

```python
y_train.value_counts()
```

```python
from collections import Counter

from imblearn.under_sampling import NearMiss
ns = NearMiss(sampling_strategy=0.8)

X_train_ns,y_train_ns = ns.fit_resample(X_train,y_train)

print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes before fit {}".format(Counter(y_train_ns)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train_ns,y_train_ns)
```

```python
y_pred = rf.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

#### Over Sampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
os = RandomOverSampler(sampling_strategy=0.5)

X_train_os,y_train_os = os.fit_resample(X_train,y_train)

print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes before fit {}".format(Counter(y_train_os)))

## In oversampling we increase the values in class where we have low value....
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train_os,y_train_os)

y_pred = rf.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

#### SMOTETomek

```python
from imblearn.combine import SMOTETomek
```

```python
os = SMOTETomek(sampling_strategy=0.5) ## SMOTETomek uses the both concept of undersamplng and oversampling...

X_train_os,y_train_os = os.fit_resample(X_train,y_train)
print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes before fit {}".format(Counter(y_train_os)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train_os,y_train_os)

y_pred = rf.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
